```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("soil_moisture_calibration_template.csv")

df["wet_weight_g"] = pd.to_numeric(df["wet_weight_g"], errors="coerce")
df["dry_weight_g"] = pd.to_numeric(df["dry_weight_g"], errors="coerce")

df = df.dropna(subset=["wet_weight_g", "dry_weight_g"])

df["moisture_percent"] = ((df["wet_weight_g"] - df["dry_weight_g"]) / df["dry_weight_g"]) * 100

x = df["soil_moisture_reading"]
y = df["moisture_percent"]

coeffs = np.polyfit(x, y, 1)
best_fit = np.poly1d(coeffs)

plt.figure()
plt.scatter(x, y)
plt.plot(x, best_fit(x))

plt.xlabel("Soil Moisture Sensor Reading")
plt.ylabel("Moisture Percentage (%)")
plt.title("Calibration Curve")

plt.show()

```
